```python
#ARIMA for AAPL
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy
from statsmodels.tsa.seasonal import seasonal_decompose
# create a differenced series
airline = pd.read_csv('C:/Users/baps/online class/AAPL1.csv',index_col ='Date')
print(airline.head())


# seasonal difference
X = airline[1:].values
#print(X)
result = seasonal_decompose(airline['Adj. Close'],model ='multiplicative',period=np.all(X))

result.plot()
plt.show()

p=d=q=range(0,2)
import itertools
pdq=list(itertools.product(p,d,q))
seasonal_pdq=[(x[0],x[1],x[2],12) for x in pdq]
print("Examples of Parameter for SARIMA:")
print("SARIMAX{}x{}".format(pdq[1],seasonal_pdq[1]))
print("SARIMAX{}x{}".format(pdq[1],seasonal_pdq[2]))
print("SARIMAX{}x{}".format(pdq[2],seasonal_pdq[3]))
for p in pdq:
    for ps in seasonal_pdq:
        try:
            mod=sm.tsa.statespace.SARIMAX(airline,order=p,seasonal_order=ps,enforce_stationarity=False,enforce_invertibility=False)
            result=mod.fit()
            print("ARIMAX{}x{}12".format(p,ps),result.aic)
        except:
            continue

train= airline.iloc[:len(airline)-12]
test= airline.iloc[len(airline)-12:]
mod=sm.tsa.statespace.SARIMAX(train,order=(0,1,1),seasonal_order=(1,1,1,12),enforce_stationarity=False,
enforce_invertibility=False)
result=mod.fit()

print(result.summary().tables[1])

result.plot_diagnostics(figsize=(16,8))
plt.show()

start=len(train)
end=len(train)+len(test)-1

predictions= result.predict(start,end,typ='levels').rename("predictions")

predictions.plot(legend=True)
test['Adj. Close'].plot(legend=True)
plt.show()


#forecasting:

start=len(train)
end=(len(train)+len(test)-1)+3*12

forecast= result.predict(start,end,typ='levels').rename("Forecasting")

airline['Adj. Close'].plot(figsize=(12,6),legend=True)
forecast.plot(legend=True)

plt.show()
```
